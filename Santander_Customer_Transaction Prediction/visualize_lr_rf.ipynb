```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def load_data(file_path):
    x_data, train_result, val_result = [], [], []
    with open(file_path, 'r') as f:
        for line in f.readlines():
            items = line.split("|")
            # x_data_curr = float(items[0][6:]) # or int(), 这里要注意根据具体情况切片
            x_data_curr = items[0][13:]
            train_result_curr = float(items[1][9:])
            val_result_curr = float(items[2][9:])
            x_data.append(x_data_curr)
            train_result.append(train_result_curr)
            val_result.append(val_result_curr)
    return x_data, train_result, val_result
```

```python
def plot_train_val(x_data, train_result, val_result, title, save_path=None, ylim_tr=(0.9850,0.9980), ylim_cv=(0.8900,0.9000)):
    # x_data = np.array(x_data)
    x_value = list(range(len(x_data)))
    train_result = np.array(train_result)
    val_result = np.array(val_result)
    
    
    plt.figure(figsize=(20,8))
    plt.subplot(121)
    plt.bar(x_value, train_result, color = "red", alpha = 0.7)
    plt.title("Train")
    plt.xlabel(title)
    plt.ylabel("roc auc")
    plt.xticks(x_value, x_data)
    plt.ylim(ylim_tr)
    plt.xlim(-1,4)
    plt.grid()
    
    plt.subplot(122)
    plt.bar(x_value, val_result, color = "blue", alpha = 0.7) # plot 折线，bar 柱形图
    plt.title("Validation")
    plt.xlabel(title)
    plt.ylim(ylim_cv)
    plt.xticks(x_value, x_data)
    plt.ylabel("roc auc")
    plt.xlim(-1,4)
    plt.grid()
    
    if save_path is not None:
        plt.savefig(save_path)
    plt.show()
```

```python
# rf round 1: number of estimators
file_path = "result/rf_001_n_estimator.txt"
x_data, train_result, val_result = load_data(file_path)
title = "n_estimators"
save_path = "figure/rf_001_n_estimator.png"
plot_train_val(x_data, 
               train_result, 
               val_result, 
               title, 
               save_path=save_path, 
               ylim_tr=(0.95,1.01),
               ylim_cv=(0.75,0.9))
```
