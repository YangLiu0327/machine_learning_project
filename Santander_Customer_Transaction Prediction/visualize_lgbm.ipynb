```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plot_train_val(x_data, train_result, val_result, title, save_path=None):
    x_data = np.array(x_data)
    train_result = np.array(train_result)
    val_result = np.array(val_result)
    
    plt.figure(figsize=(20,8))
    plt.subplot(121)
    plt.bar(x_data, train_result, color = "red", width=0.4, alpha = 0.7)
    plt.title("Train")
    plt.xlabel(title)
    plt.ylabel("roc auc")
    plt.xticks(x_data)
    plt.ylim(0.9570,0.9835)
    plt.grid()
    
    plt.subplot(122)
    plt.bar(x_data, val_result, color = "blue",width=0.4,  alpha = 0.7)
    plt.title("Validation")
    plt.xlabel(title)
    plt.ylim(0.7600,0.9040)
    plt.xticks(x_data)
    plt.ylabel("roc auc")
    plt.grid()
    
    if save_path is not None:
        plt.savefig(save_path)
    plt.show()
```

```python
# max_depth

title = "max_depth"
x_data = list(range(3,26,2))
train_result = [0.9926,0.9941,0.9952,0.9900,0.9909,0.9922,0.9921,0.9921,0.9921,0.9921,0.9921,0.9921]
val_result = [0.8970,0.8921,0.8947,0.8959,0.8947,0.8944,0.8946,0.8946,0.8946,0.8946,0.8946,0.8946]
plot_train_val(x_data, train_result, val_result, title, save_path="max_depth.png")
```
